```python
def moving_average(a, n=7):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n


a = np.random.normal(0,1,(1000,))
b = moving_average(a, 15)
plt.plot(a)
plt.plot(b)
plt.show()
```

```python
def gompertz(deltat, n, r, k):
    '''
    Curva de crecimiento de Gomperz, caso especial de la GL == 0
    '''
    if (k > 1e-6):
        return deltat * r * n * np.log(k / n)
    else:
        return deltat * r * n * np.log(1e-6 / n)

def rk4g(dx, y, r, k):
    #integrador Runge-Kutta 4
    k1 = gompertz(dx, y, r, k)
    k2 = gompertz(dx, y + k1 / 2, r, k)
    k3 = gompertz(dx, y + k2 / 2, r, k)
    k4 = gompertz(dx, y + k3, r, k)
    return y + (k1 + 2 * k2 + 2 * k3 + k4) / 6.0
```

```python
n = np.zeros((1000,))
m = np.zeros((1000,))

m[0] = 200
n[0] = 0.1
for i in range (1, n.shape[0]):
    n[i] = n[i-1] + gompertz(deltat, n[i-1], r, k)
    m[i] = m[i-1] + gompertz(deltat, m[i-1], r, k)

plt.plot(n)
plt.plot(m)
```

```python
## Ahora hacemos con rk4g y comparamos

deltat = 0.1
nr = np.zeros((1000,))
mr = np.zeros((1000,))
r = 0.1
k = 100
nr[0] = 0.1
mr[0] = 200

for i in range (1, nr.shape[0]):
    nr[i] = rk4g(deltat, n[i-1], r, k)
    mr[i] = rk4g(deltat, m[i-1], r, k)
    

plt.plot(n+10, label="gomperz n0=0.1 (+10)")
plt.plot(m+10, label="gomperz n0=200 (+10)")
plt.plot(nr, label="rk4g n0=0.1")
plt.plot(mr, label="rk4g n0=200")

plt.legend()
```
